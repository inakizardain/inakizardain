## Introduction

### Since 2012 plenty of states of the USA have legalized the recreational use and possession of cannabis as a strategy to damage narcotrafic groups and to supply consumers with high quality products. Nowadays 36% (18 out of 50 states) have liberalized cannabis consumption.

```python
def col_change (df):
  """
  Rename column names and change all values to float

  Args:
    df (DataFrame): the DataFrame to rename and change data type

  Returns:
    DataFrame
  """
  df.rename(columns={df.columns[0]: 'State', df.columns[1]:'TEOGS', df.columns[2]:'TCPO', df.columns[3]:'TEIGS', df.columns[4]:'TCPI', df.columns[5]:'TCPOI', df.columns[6]:'BPM', df.columns[7]:'NOA', df.columns[8]:'ASV', df.columns[9]:'WS'}, inplace=True)
  for name in df.columns[1:]:
    df[name]=df[name].apply(lambda x: x.replace(',', ''))
    df[name]=df[name].apply(lambda x: x.replace('$', ''))
    df[name]=df[name].astype(float)
  return df.columns
```

```python
def column_df (df, column_name):
  """
  Create a DataFrame that groups columns with the same column name

  Args:
    df (DataFrame): DataFrame to split in groups
    column_name (string): Name of the column without suffixes that we want to group

  Returns:
    DataFrame
  """
  lista=[col for col in df.columns if column_name in col[:-4] and len(col)==len(column_name)+4]
  lista.insert(0, 'State')
  return df[lista]
 
```

```python
def plots(df, state_names, legal_year_cols=None):
    """
    Lineplot of all columns from a DataFrame for multiple states with lines representing their legalization years

    Args:
      df (DataFrame): DataFrame to be plotted
      state_names (list): list of state names
      legal_year_cols (list): list of column names representing the years when cannabis was legalized for each state
    
    Returns:
      Lineplots for multiple states in the same row
    """
    num_plots = len(state_names)
    fig, axes = plt.subplots(1, num_plots, figsize=(num_plots*8, 6))

    for i in range(num_plots):
        var = df[df['State'] == state_names[i]]
        var_lim = [year for year in var.columns[1:]]
        sns.lineplot(x=var_lim, y=[var[j].values[0] for j in var_lim], ax=axes[i])
        axes[i].set_title(state_names[i])
        axes[i].set_xticklabels(var_lim, rotation=90)

        if legal_year_cols:
            legal_year_col = legal_year_cols[i]
            axes[i].axvline(x=legal_year_col, linestyle='--', color='red', label=f'Legal year {legal_year_col[-4:]}')
            axes[i].legend()

    plt.tight_layout()
    plt.show()
```

```python
def over_plots(df, state_names):
  """
  Lineplot of all columns from a DataFrame by a group of states 

  Args:
    df (DataFrame): DataFrame to be plotted
    state_name (list): list with the names of the states

  Returns:
    Lineplot  
  """
  fig, ax = plt.subplots(figsize=(8,4))
  colors = sns.color_palette('husl', len(state_names))
  for i, state_name in enumerate(state_names):
    var = df[df['State'] == state_name]
    var_lim = [year for year in var.columns[1:]]
    sns.lineplot(x=var_lim, y=[var[z].values[0] for z in var_lim], ax=ax, alpha=1, label=state_name, color=colors[i])
  plt.xticks(rotation=90)
  ax.legend() #ax is needed, at the begining and inside of the sns.barplot() so that the figsize can be different and we can add a legend with the name of the city analyzed
  plt.show()
```

## Objective 2: Analyze the trend of TCPOI, ASV and WS for 3 outlier states after their respective cannabis librealizations
### As we will see, the states that are considered to be outliers in the three values to be studied are California, Kentucky, Michigan, Georgia, Oregon and Indiana. We will only analyze California, Oregon and Michigan

```python
def best_version(df, top_outliers=True, low_outliers=True):
    """
    Dictionary with the name of the state that is an outlier for 1 or more years and the number of
    times from 2011-2021 that state has been an outlier

    Args:
    df (DataFrame): DataFrame to be analyzed
    top_outliers (boolean): If True, find top outliers; Else, don't find top outliers
    low_outliers (boolean): If True, find low outliers; Else, don't find low outliers

    Results:
    Dictionary
    """
    dict_outliers = {}
    for col in df.columns[1:]:
        q3 = df[col].quantile(.75)
        q1 = df[col].quantile(.25)
        IQR = q3 - q1
        top_range = q3 + (1.5 * IQR)
        low_range = q1 - (1.5 * IQR)
        if top_outliers:
            df_top = df.loc[df[col] > top_range]
            df_top_list = df_top['State'].tolist()
            for element in df_top_list:
                if element in dict_outliers.keys():
                    dict_outliers[element] += 1
                else:
                    dict_outliers[element] = 1
        if low_outliers:
            df_low = df.loc[df[col] < low_range]
            df_low_list = df_low['State'].tolist()
            for element in df_low_list:
                if element in dict_outliers.keys():
                    dict_outliers[element] += 1
                else:
                    dict_outliers[element] = 1
    return dict_outliers
```
